# Trip Distance
Calculate the distance between pickup and dropoff points using the Haversine formula.

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c 
    return distance

new_df['trip_distance_km'] = new_df.apply(lambda row: haversine(row['pickup_latitude'], row['pickup_longitude'], 
                                                        row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
print(new_df[['pickup_latitude', 'pickup_longitude', 'dropoff_latitude', 'dropoff_longitude', 'trip_distance_km']])
```

# Handling Outliers


```python
import pandas as pd

# Outlier bounds for longitude and latitude (based on NYC)
longitude_min, longitude_max = -74.26, -73.70
latitude_min, latitude_max = 40.49, 40.92

# Detecting outliers in pickup and dropoff coordinates
outliers_geo = new_df[
    (new_df['pickup_longitude'] < longitude_min) | (new_df['pickup_longitude'] > longitude_max) |
    (new_df['pickup_latitude'] < latitude_min) | (new_df['pickup_latitude'] > latitude_max) |
    (new_df['dropoff_longitude'] < longitude_min) | (new_df['dropoff_longitude'] > longitude_max) |
    (new_df['dropoff_latitude'] < latitude_min) | (new_df['dropoff_latitude'] > latitude_max)
]

# Remove outliers based on geographic coordinates
new_df_no_geo_outliers = new_df.drop(outliers_geo.index)

Q1_distance = new_df_no_geo_outliers['trip_distance_km'].quantile(0.25)
Q3_distance = new_df_no_geo_outliers['trip_distance_km'].quantile(0.75)
IQR_distance = Q3_distance - Q1_distance

lower_bound_distance = Q1_distance - 1.5 * IQR_distance
upper_bound_distance = Q3_distance + 1.5 * IQR_distance

outliers_distance = new_df_no_geo_outliers[
    (new_df_no_geo_outliers['trip_distance_km'] < lower_bound_distance) | 
    (new_df_no_geo_outliers['trip_distance_km'] > upper_bound_distance)
]

new_df_final_no_outliers = new_df_no_geo_outliers.drop(outliers_distance.index)
```

```python
new_df['distance_band'] = pd.cut(new_df['trip_distance_km'], bins=[0, 3, 15, np.inf], labels=['short', 'medium', 'long'])

def remove_outliers_by_band(new_df, column, group_col):
    def iqr_outliers(group):
        Q1 = group[column].quantile(0.25)
        Q3 = group[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return group[(group[column] >= lower_bound) & (group[column] <= upper_bound)]
    
    return new_df.groupby(group_col, group_keys=False).apply(iqr_outliers)

# Step 1: Remove outliers based on fare amount within distance bands
df_no_outliers_by_distance = remove_outliers_by_band(new_df, 'fare_amount', 'distance_band')

# Step 2: Remove outliers based on fare amount within passenger count groups
df_no_outliers_by_passenger = remove_outliers_by_band(df_no_outliers_by_distance, 'fare_amount', 'passenger_count')

# Step 3: Calculate fare per kilometer and filter out unreasonable values (based on IQR bounds for fare_per_km)
df_no_outliers_by_passenger['fare_per_km'] = df_no_outliers_by_passenger['fare_amount'] / df_no_outliers_by_passenger['trip_distance_km']

lower_bound_fare_per_km = 0  # Fare per km cannot be negative
upper_bound_fare_per_km = 7.251726409772042  # Upper bound from IQR calculation

# Step 4: Filter the dataset based on the fare_per_km bounds
df_fare_per_km_reasonable = df_no_outliers_by_passenger[(df_no_outliers_by_passenger['fare_per_km'] >= lower_bound_fare_per_km) & 
                                                        (df_no_outliers_by_passenger['fare_per_km'] <= upper_bound_fare_per_km)]

# Step 5: Remove outliers based on trip distance (based on IQR bounds for trip distance)
lower_bound_trip_distance = 0  # Trip distance cannot be negative
upper_bound_trip_distance = 7.901128252810093  # Upper bound from IQR calculation

df_final_cleaned = df_fare_per_km_reasonable[(df_fare_per_km_reasonable['trip_distance_km'] >= lower_bound_trip_distance) &
                                             (df_fare_per_km_reasonable['trip_distance_km'] <= upper_bound_trip_distance)]

print(df_final_cleaned.head())
```
